```python
port=data.drop(['Ship Mode','Country','City','State','Postal Code','Region'],axis=1)
```

```python
port.head()
```

```python
port.tail()
```

```python
corelation=port.corr()
```

```python
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,
            annot=True)
```

```python
sns.pairplot(port)
```

```python
sns.relplot(x='Sales',y='Quantity',hue='Discount',data=port)
```

```python
sns.relplot(x='Sales',y='Quantity',hue='Profit',data=port)
```

```python
sns.distplot(port['Sales'],bins=5)
```

```python
sns.distplot(port['Quantity'])
```

```python
sns.catplot(x='Quantity',kind='box',data=port)
```
